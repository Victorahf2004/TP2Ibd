```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv('MICRODADOS_ENEM_2021.csv', encoding='latin-1', sep = ';')
```

```python
data = pd.read_csv('rendimento_domiciliar.csv', encoding='latin-1', sep = ';')
data = data.rename(columns={'ï»¿uf': 'SG_UF_PROVA'})
data
```

```python
colunas_especificas = ['NU_INSCRICAO','TP_SEXO', 'TP_COR_RACA', 'SG_UF_PROVA', 'NU_NOTA_REDACAO',  'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_MT', 'NU_NOTA_LC']
df = df.dropna(subset=colunas_especificas)
df = df[colunas_especificas]
```

```python
df
```

```python
# Colunas específicas para calcular a média
notas = ['SG_UF_PROVA', 'NU_NOTA_REDACAO', 'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_MT', 'NU_NOTA_LC']

# Média das notas específicas por estado
media_por_estado = df.groupby('SG_UF_PROVA')[notas].mean().reset_index()

media_por_estado
```

```python
notas = ['SG_UF_PROVA', 'NU_NOTA_REDACAO', 'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_MT', 'NU_NOTA_LC']

media_por_estado = df.groupby('SG_UF_PROVA')[notas].mean()

print(media_por_estado)
print()

# Estado com a maior e menor nota para cada coluna
notas2 = ['NU_NOTA_REDACAO', 'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_MT', 'NU_NOTA_LC']

for coluna in notas2:
    estado_maior_nota = media_por_estado[coluna].idxmax()
    estado_menor_nota = media_por_estado[coluna].idxmin()

    print(f"Coluna: {coluna}")
    print(f"Estado com a maior nota: {estado_maior_nota}, Nota: {media_por_estado[coluna][estado_maior_nota]}")
    print(f"Estado com a menor nota: {estado_menor_nota}, Nota: {media_por_estado[coluna][estado_menor_nota]}")
```

```python
notas = ['SG_UF_PROVA', 'NU_NOTA_REDACAO', 'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_MT', 'NU_NOTA_LC']

media_por_estado = df.groupby('SG_UF_PROVA')[notas].mean()

print(media_por_estado)
print()

notas2 = ['NU_NOTA_REDACAO', 'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_MT', 'NU_NOTA_LC']

for coluna in notas2:
    estado_maior_nota = df.loc[df[coluna] == df[coluna].max(), 'SG_UF_PROVA'].iloc[0]

    print(f"Coluna: {coluna}")
    print(f"Estado com a maior nota: {estado_maior_nota}, Nota: {df[coluna].max()}")
```

```python
notas = ['SG_UF_PROVA', 'NU_NOTA_REDACAO', 'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_MT', 'NU_NOTA_LC']

media_por_estado = df.groupby('SG_UF_PROVA')[notas].mean()

print(media_por_estado)
print()

notas2 = ['NU_NOTA_REDACAO', 'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_MT', 'NU_NOTA_LC']

for coluna in notas2:
    maiores_notas = df.nlargest(10, coluna)

    print(f"Coluna: {coluna}")
    for _, row in maiores_notas.iterrows():
        estado = row['SG_UF_PROVA']
        nota = row[coluna]
        print(f"Estado: {estado}, Nota: {nota}")
    print()
```

```python
import pandas as pd

# Criar mapeamento de estados para regiões
regioes = {
    'Norte': ['AM', 'RR', 'AP', 'PA', 'TO', 'RO', 'AC'],
    'Nordeste': ['MA', 'PI', 'CE', 'RN', 'PE', 'PB', 'SE', 'AL', 'BA'],
    'Sudeste': ['MG', 'ES', 'RJ', 'SP'],
    'Sul': ['PR', 'SC', 'RS'],
    'Centro-Oeste': ['MS', 'MT', 'GO', 'DF']
}

# Adicionar coluna de região ao DataFrame
df['Regiao'] = df['SG_UF_PROVA'].map({estado: regiao for regiao, estados in regioes.items() for estado in estados})

# Calcular média por região
media_por_regiao = df.groupby('Regiao')[notas].mean()

print(media_por_regiao)
print()

for coluna in notas2:
    regiao_maior_media = media_por_regiao[coluna].idxmax()
    regiao_menor_media = media_por_regiao[coluna].idxmin()

    print(f"Coluna: {coluna}")
    print(f"Região com maior média: {regiao_maior_media}, Média: {media_por_regiao.loc[regiao_maior_media, coluna]}")
    print(f"Região com menor média: {regiao_menor_media}, Média: {media_por_regiao.loc[regiao_menor_media, coluna]}")
    print()

```
